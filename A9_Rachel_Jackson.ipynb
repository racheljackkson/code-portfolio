```
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
import numpy as np
df = pd.read_csv('survey-1.csv', index_col=0, na_values=[''])
```

```
df.dtypes
```

```
sns.set()
plt.figure()

xyz = sns.barplot(data=df, x="Major", y="LikeMath", hue="Programming", palette=['green', 'red'], saturation=0.5)
xyz.set_title('Rating of Math by Major and Programming Experience')
xyz.set_ylabel('Rating of Math (1-5)')
xyz.set_xlabel(None)
xyz.set_xticklabels(df["Major"].unique().astype(str), rotation='vertical')
plt.show()
```

```
sns.set_color_codes(palette='bright')
sns.distplot(df.Siblings, color="y")
plt.xlim(1,8)
plt.title("Distribution of No. of Siblings")
```

```
sns.set_theme(style="ticks")

f, bx = plt.subplots(figsize=(8, 7))

sns.boxplot(x="LikeMath", y="Sports", data=df, width=.6, palette="vlag")
sns.stripplot(x="LikeMath", y="Sports", data=df,
              size=4, color=".3", linewidth=0)

bx.xaxis.grid(True)
bx.set_title('Rating of Math by Sports Fan')
bx.set_xlabel('Rating of Math')
bx.set_ylabel('Sports Fan')
sns.despine(trim=True, left=True)
```

```
sns.set_theme(style="whitegrid", palette="muted")
ax = sns.swarmplot(x="InstructorGreatness", y="LikeMath", data=df, hue='Python')
ax.set_title('Instructor Greatness vs Math Rating given Python Experience')
plt.tight_layout()
ax.set_xlabel('Rating of Instructor Greatness')
ax.set_ylabel('Rating of Math')
```
